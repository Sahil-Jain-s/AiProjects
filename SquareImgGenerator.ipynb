```python
img = Image.new('RGB', (200,200), color = '#FF6633')
draw = ImageDraw.Draw(img)
for i in range(10):
  stopx,stopy=200,200
  x1=random.randrange(0,stopx/2)
  y1=random.randrange(0,stopy/2)
  x2=random.randrange(stopx/2,stopx)
  y2=random.randrange(stopy/2,stopy)
  start_angle=random.randrange(0,180)
  end_angle=random.randrange(90,180)+start_angle
  xy=[x1,y1,x2,y2]
  start=15
  end=600
  draw.ellipse(xy, fill=colors_array[i], width=0)
img.save(f'fimg.png')
```

```python
img = Image.new('RGB', (200,200), color = '#FF6633')
draw = ImageDraw.Draw(img)
max=5
stopx,stopy=200,200
k=1
radius=20
margin=20
for j in range(max):
  x1=margin+j*stopx/max
  x2=x1+radius
  for i in range(max):
    y1=margin+i*stopy/max
    y2=y1+radius
    xy=[x1,y1,x2,y2]
    draw.ellipse(xy, fill=colors_array[k], width=0)
    k=k+1

img.save(f'fimg.png')
```
